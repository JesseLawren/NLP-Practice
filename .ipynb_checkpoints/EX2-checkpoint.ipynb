```python
import requests
websiteGE = 'https://www.gutenberg.org/cache/epub/4363/pg4363.txt'
siteOne = requests.get(websiteGE)
websitePlato = 'https://www.gutenberg.org/cache/epub/1497/pg1497.txt'
siteTwo = requests.get(websitePlato)
```

```python
siteOneText = siteOne.text
siteTwoText = siteTwo.text

clean_siteOneText = siteOneText.replace('\r\n', ' ').replace('\n' , ' ').replace('\r', ' ').replace("it's", "it is").replace("they're", "they are").replace("isn't", "is not").replace("they'll", "they will").lower()
clean_siteTwoText = siteTwoText.replace('\r\n', ' ').replace('\n' , ' ').replace('\r', ' ').replace("it's", "it is").replace("they're", "they are").replace("isn't", "is not").replace("they'll", "they will").lower()
```

```python
print(clean_siteOneText[1000:1500])
print('\n')
print(clean_siteTwoText[1000:1500])
```

```python
import re
def find_sentences_with_pattern(text, pattern):
    sentences = re.split(r'[.!?]', text)
    matches = [sentence.strip() + '.' for sentence in sentences if re.search(pattern, sentence)]
    return matches

pattern = r'\bphiloso[a-z]+\b'

matchesOne = find_sentences_with_pattern(clean_siteOneText, pattern)
matchesTwo = find_sentences_with_pattern(clean_siteTwoText, pattern)
```

```python
print('\n')
print('We found ' + str(len(matchesOne)) + ' matches in the first book...')
print(str(matchesOne))
print('--------------------------------')
print('\n')
print('We found ' + str(len(matchesTwo)) + ' matches in the second book...')
print(str(matchesTwo))
```

```python
def tokenize_and_print_last_word(sentences):
    all_words = []
    for sentence in sentences:
        tokens = re.findall(r'\w+', sentence)
        all_words.extend(tokens)
    return all_words

wordsOne = tokenize_and_print_last_word(matchesOne)
wordsTwo = tokenize_and_print_last_word(matchesTwo)
```

```python
print(wordsOne)
print(wordsOne[-1])
print('---------------------------------------------')
print(wordsTwo)
print(wordsTwo[-1])
```

```python
import matplotlib.pyplot as plt
from collections import Counter

def plot_common_words(words, title, n = 10):
    word_freq = Counter(words)
    common_words, counts = zip(*word_freq.most_common(n))

    plt.bar(common_words, counts)
    plt.title(title)
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation = 45)
    plt.show()

```

```python
plot_common_words(wordsOne, 'Most Common Words in Book One')
plot_common_words(wordsTwo, 'Most Common Words in Book Two')
```
