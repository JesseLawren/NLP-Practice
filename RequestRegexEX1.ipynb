```python
import requests
website = 'https://www.gutenberg.org/files/4280/4280-0.txt'
site = requests.get(website)
```

```python
import re
regular_expression = '(p|P)hiloso[a-z]+'
text = site.text
results = [{"indices":m.span(), "match":m.group() } for m in re.finditer(regular_expression, text)]
matches = []; [matches.append(m['match']) for m in results]
```

```python
#Without regex syntax
print('We found ' + str(len(matches)) + ' matches...')
print(str(results))
```

```python
#With regex syntax
print('We found ' + str(len(matches)) + ' matches...')
print(str(results))
```
